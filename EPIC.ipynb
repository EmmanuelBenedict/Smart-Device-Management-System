```python
class SmartDevice:                                                         #Parent class
    def __init__(self,name,device_id,power_status):
        self.name = name
        self.__device_id = device_id
        self.__power_status =power_status

    def turn_on(self):                                                     #turn on method
        self.__power_status = "ON"
        print(f"{self.name} is turned on")

    def turn_off(self):                                                             #turn off method
        self.__power_status = "OFF"
        print(f"{self.name} is turned off")

    def display_info(self):                                                       #display method
        print(f"{self.name} with device id:{self.__device_id} is {self.__power_status}")

    def get_device_id(self):                                                      #getter
        return self.__device_id

    def get_power_status(self):                                                  #getter
        return self.__power_status

    def set_device_id(self,new_id):                                             #setter
        if new_id == "":
            print("device id cannot be empty")
        else:
            self.__device_id = new_id

    def set_power_status(self,new_power_status):                              #setter
        if new_power_status == "ON":
            self.__power_status = "ON"
        elif new_power_status == "OFF":
            self.__power_status = "OFF"
        else:
            print("Invalid power status")


class TemperatureSensor(SmartDevice):                                    #child class temperature sensor
    def __init__(self,name,device_id,temperature):
        super().__init__(name,device_id,"OFF")
        self.__temperature = temperature

    def get_temperature(self):                                           #temperature getter
        return self.__temperature

    def set_temperature(self,new_temperature):
        self.__temperature = new_temperature

    def read_temperature(self):
        print(f"{self.name} is {self.__temperature} degrees Celsius")

    def display_info(self):
        super().display_info()
        print(f" Temperature : {self.__temperature} degrees Celsius")


class SecurityCamera(SmartDevice):
    def __init__(self,name,device_id):
        super().__init__(name,device_id,"OFF")
        self.__recording_status = "OFF"

    def get_recording_status(self):
        return self.__recording_status

    def start_recording(self):
        self.__recording_status = "ON"
        print(f"{self.name} is recording")

    def stop_recording(self):
        self.__recording_status = "OFF"
        print(f"{self.name} has stopped recording")

    def display_info(self):
        super().display_info()
        print(f" Recording Status : {self.__recording_status}")


class SmartLight(SmartDevice):
    min_brightness = 0
    max_brightness = 100
    def __init__(self,name,device_id):
        super().__init__(name,device_id,"OFF")
        self.__brightness =50

    def get_brightness(self):
        return self.__brightness

    def set_brightness(self,new_brightness):
        if self.min_brightness <= new_brightness <= self.max_brightness:
            self.__brightness = new_brightness
        else:
            print("brightness out of range")



    def increase_brightness(self,amount):
        if self.__brightness + amount <= self.max_brightness:
            self.__brightness += amount
        else:
            self.__brightness = self.max_brightness
        print(f"brightness = {self.__brightness}")

    def decrease_brightness(self,amount):
        if self.__brightness - amount  >= self.min_brightness:
            self.__brightness -= amount
        else:
            self.__brightness = self.min_brightness
        print(f"brightness = {self.__brightness}")

    def display_info(self):
        super().display_info()
        print(f" Brightness : {self.__brightness}")

kitchen_temp=TemperatureSensor("Temperature Sensor",55454,25)
backyard_camera=SecurityCamera("smart camera",5657)
living_room_light=SmartLight("Smart light",55454)



while True:
    print("1. Display device information")
    print("2. Turn on")
    print("3. Turn off")
    print("4. Read temperature")
    print("5. Adjust brightness")
    print("6. Turn Start recording")
    print("7. Exit")

    choice=int(input("Enter your choice:"))
    if choice==1:
        kitchen_temp.display_info()
        backyard_camera.display_info()
        living_room_light.display_info()

    elif choice==2:
        kitchen_temp.turn_on()
        backyard_camera.turn_on()
        living_room_light.turn_on()

    elif choice==3:
        kitchen_temp.turn_off()
        backyard_camera.turn_off()
        living_room_light.turn_off()

    elif choice==4:
        kitchen_temp.read_temperature()

    elif choice==5:
        amount=int(input("Enter the amount you want to increase:"))
        living_room_light.increase_brightness(amount)

    elif choice==6:
        backyard_camera.start_recording()

    elif choice==7:
        print("Exit")
        break

    else:
        print("Invalid choice. please try again")
```
